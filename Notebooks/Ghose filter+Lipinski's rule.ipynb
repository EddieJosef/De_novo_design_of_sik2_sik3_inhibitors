**Ghose filter:**

* log P  [−0.4 ; +5.6]
* Molecular weight [180 ; 480]
* Number of atoms [20 ; 70]
* Molar refractivity [40 ; 130]
* ≤10 rotatable bonds
* Polar surface area ≤ 140

**Lipinski's rule:**
* No more than 5 hydrogen bond donors
* No more than 10 hydrogen bond acceptors
* A molecular mass less than 500 daltons
* A calculated octanol-water partition coefficient (Clog P) that does not exceed 5

```python
# data_frame = PandasTools.LoadSDF("Asinex_BioDesign.sdf")
# frame = data_frame.ROMol
# Load SDF file
suppl = Chem.SDMolSupplier('Enamine_MiniFrag.sdf')

# Initialize a list to store SMILES strings
smiles = []

# Iterate over molecules in the SDF file
for mol in suppl:
    # Check if the molecule is valid
    if mol:
        # Convert molecule to SMILES
        smile = Chem.MolToSmiles(mol)
        smiles.append(smile)

# Print or use the SMILES list as needed
print(smiles)
```

```python
smiles_canonic = []
for smile in smiles:
    mol = Chem.MolFromSmiles(smile)
    if mol is not None:
        smi = Chem.MolToSmiles(mol)
        smiles_canonic.append(smi)
    else:
        print(f"Unable to parse SMILES: {smile}")
```

```python
molecules = []

for smile in smiles_canonic:
    mol = Chem.MolFromSmiles(smile)
    molecules.append(mol)
```

```python
molecular_descriptors = {
    "NumHAcceptors": Descriptors.NumHAcceptors, # No more than 10 hydrogen bond acceptors
    "NumHDonors": Descriptors.NumHDonors, # No more than 5 hydrogen bond donors
    "NumRotatableBonds": Descriptors.NumRotatableBonds, # ≤10 rotatable bonds
    "MW": Descriptors.MolWt, # Molecular weight [180 ; 480]
    "LogP": Descriptors.MolLogP, # log P [−0.4 ; +5.6]
    "TPSA": Descriptors.TPSA # Polar surface area ≤ 140
}

def mol_dsc_calc(mols):
    return DataFrame({k: f(m) for k, f in molecular_descriptors.items()}
             for m in mols)

descriptors_transformer = FunctionTransformer(mol_dsc_calc, validate=False)
```

```python
dataframe_calculcated = descriptors_transformer.transform(molecules)
```

```python
smiles_column = pd.Series(smiles_canonic)
```

```python
dataset = pd.concat([smiles_column.rename('smiles'), dataframe_calculcated], axis=1)
```

```python
# Conditions for filtering
condition_acceptors = dataset['NumHAcceptors'] <= 10
condition_donors = dataset['NumHDonors'] <= 5
condition_rotatable_bonds = dataset['NumRotatableBonds'] <= 10
condition_mw = (dataset['MW'] >= 180) & (dataset['MW'] <= 480)
condition_logP = (dataset['LogP'] >= -0.4) & (dataset['LogP'] <= 5.6)
condition_tpsa = dataset['TPSA'] <= 140

# Combine conditions with 'and' (&) to filter rows that meet all criteria
filtered_dataset = dataset[
    condition_acceptors &
    condition_donors &
    condition_rotatable_bonds &
    condition_mw &
    condition_logP &
    condition_tpsa
]

# Print the filtered dataset
```
